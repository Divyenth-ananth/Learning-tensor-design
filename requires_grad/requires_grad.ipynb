```python
import torch
import matplotlib.pyplot as plt
```

```python
# Ground truth y
a = 2.0
b = -3.0
c = 1.5
x = torch.linspace(-10, 10, 1000)
y = a * x**2 + b * x + c
```

```python
# Initial guesses
a1 = torch.randn(1, requires_grad=True)
b1 = torch.randn(1, requires_grad=True)
c1 = torch.randn(1, requires_grad=True)
alpha = 1e-4
for epoch in range(150):  # doesn't even get close for 20-30 epochs
  y_pred = a1*x**2 + b1*x + c1
  loss = torch.mean((y_pred - y)**2)

  a1.grad = None
  b1.grad = None
  c1.grad = None

  loss.backward()

  with torch.no_grad():
     a1 -= alpha * a1.grad
     b1 -= alpha * b1.grad
     c1 -= alpha * c1.grad

  print(f"Epoch {epoch+1}: a1={a1.item():.4f}, b1={b1.item():.4f}, c1={c1.item():.4f}, loss={loss.item():.2f}")

```
